```python
### Workshop 001 - WebCrawling

import requests
from bs4 import BeautifulSoup

url = "https://www.tripadvisor.com.sg/Attraction_Review-g294264-d2439664-Reviews-Universal_Studios_Singapore-Sentosa_Island.html"
html_page = requests.get(url)
soup_page = BeautifulSoup(html_page.content, "html.parser")
containers = soup_page.findAll('div', {'class':'review-container'})

print('{:20s} {:7s} {:10s}'.format("Name", "Rating", "Reviewed Date"))
for item in containers:
    uname = str(item.find('div', {'class':'username mo'}).text)
    tContent = str(item)
    pos = tContent.find("ui_bubble_rating bubble_")
    rating = tContent[pos+24:pos+25]
    review = str(item.find('span', {'class':'ratingDate relativeDate'}).text)
    print('{:20s} {:7s} {:10s}'.format(uname, rating, review))
```

```python
### Workshop 001 - WebCrawling

import requests
import time
from bs4 import BeautifulSoup

url_prefix = 'https://www.tripadvisor.com.sg/Attraction_Review-g294264-d2439664-Reviews-'
url_suffix = 'Universal_Studios_Singapore-Sentosa_Island.html'
max_page = 2

page_no = 0

while(True):
    
    if page_no >= max_page:
        break
    else:
        if page_no == 0:
            url_page = ''
        else:
            url_page = 'or' + str(page_no) + '0-'
    
    url = url_prefix + url_page + url_suffix
    try:
        html_page = requests.get(url)
        soup_page = BeautifulSoup(html_page.content, "html.parser")
        #containers = soup_page.findAll('div', {'class':'review-container'})
        containers = soup_page.findAll('div', {'class':'review hsx_review is-multiline is-mobile inlineReviewUpdate provider0'})
        #time.sleep(1)
    except:
        containers = []
    
    for item in containers:
        review = item.find('div', {'class':'quote'}).text
        reviewText = item.find('div', {'class':'entry'}).text
        uname = item.find('span', {'class':'expand_inline scrname'}).text
        try:
            location = item.find('span', {'class':'location'}).text
        except:
            location = ''
        tContent = str(item)
        pos = tContent.find("ui_bubble_rating bubble_")
        rating = tContent[pos+24:pos+25]
        print('Review Title: ', review)
        print('Review Text: ', reviewText)
        print('Username: ', uname)
        print('Location: ', location)
        print('Rating: ', rating, 'stars\n')
    
    page_no += 1
```
