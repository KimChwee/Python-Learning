# <font color=blue> Correlation matrix of values for pandas dataframe
##### Finding correlation is key to many problems

```python
import pandas as pd
```

```python
data = {'name': ['Jack','Mason','Tanya','Jim','Amelia'],
        'tenure': [42,52,36,24,73],
        'employees': [4,24,31,2,3],
        'inventory': [25,94,57,62,70]}
df = pd.DataFrame(data, columns = ['name','tenure','employees','inventory'])
df
```

```python
df.corr()
```

```python
data = {'BattlePts': [1260,840,2150],
        'Rounds': [6,9,6],
        'Time': [4.38,6.44,4.41],
        'MaxDmg': [22.1,18.9,30.0],
        'Combos': [14,21,17],
        'Score': [2373,1465,3400]}
df = pd.DataFrame(data, columns = ['BattlePts','Rounds','Time','MaxDmg','Combos','Score'])
df
```

```python
df.corr()
```
